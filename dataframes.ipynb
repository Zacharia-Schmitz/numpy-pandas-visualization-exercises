# Dataframes Exercise

### Students Dataset

```python
# Copy the code from the lesson to create a dataframe full of student grades.

import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

stu_df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
# 1. Create a column named passing_english that indicates whether each student has a passing grade in english.

stu_df['passing_english'] = stu_df['english'] > 70

stu_df
```

```python
# 2. Sort the english grades by the passing_english column. How are duplicates handled?

stu_df.sort_values('passing_english')

# Sorted ascending by passing english, then sorted by the index number
```

```python
# 3. Sort the english grades first by passing_english and then by student name. 
# All the students that are failing english should be first, and within the students 
# that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)

stu_df.sort_values(['passing_english', 'name'])
```

```python
# 4. Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.

stu_df.sort_values(['passing_english', 'english'])
```

```python
# 5. Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.

overall_grade = (stu_df['math'] + stu_df['english'] + stu_df['reading']) / 3 

stu_df['overall_grade'] = overall_grade

stu_df.head(1)
```

### MPG Dataset

```python
# Load the mpg dataset from pydata. Read the documentation for the dataset and use it for the following questions:

import pandas as pd
from pydataset import data

data('mpg', show_doc=True)

# 1. How many rows and columns are there?

# According to the show_doc, there are 234 rows and 11 variables (columns)
```

```python
# 2. What are the data types of each column?

mpg_df = data('mpg')

mpg_df.dtypes
```

```python
# 3.1. Summarize the dataframe with .info.

mpg_df.info()
```

```python
# 3.2. Summarize the dataframe with .describe.

mpg_df.describe()
```

```python
# 4. Rename the cty column to city.

mpg_df.rename(columns={'cty': 'city'}, inplace=True)

# The inplace parameter is a boolean value that determines whether the original DataFrame is modified or a new DataFrame is returned with the modifications.

mpg_df
```

```python
# 5. Rename the hwy column to highway.

mpg_df.rename(columns={'hwy': 'highway'}, inplace=True)
```

```python
# 6. Do any cars have better city mileage than highway mileage?

mpg_df['better_city'] = mpg_df.city > mpg_df.highway

(mpg_df['city'] > mpg_df['highway']).sum()

# No results, so no
```

```python
# 7. Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.

mpg_df['mileage_differance'] = mpg_df['highway'] - mpg_df['city']

mpg_df.head(1)
```

```python
# 8. Which car (or cars) has the highest mileage difference?

mpg_df[mpg_df['mileage_differance'] == mpg_df['mileage_differance'].max()]
```

```python
# 9.1. Which compact class car has the lowest highway mileage?

mpg_df[mpg_df['class'] == 'compact'][mpg_df['highway'] == mpg_df[mpg_df['class'] == 'compact']['highway'].min()]
```

```python
# 9.2. Which compact class car has the best highway mileage?

mpg_df[mpg_df['class'] == 'compact'][mpg_df['highway'] == mpg_df[mpg_df['class'] == 'compact']['highway'].max()]
```

```python
# 10. Create a column named average_mileage that is the mean of the city and highway mileage.

mpg_df['average_mileage'] = (mpg_df['city'] + mpg_df['highway']) / 2

mpg_df.head()
```

```python
# 11.1 Which dodge car has the best average mileage?

mpg_df[mpg_df['manufacturer'] == 'dodge'].sort_values('average_mileage', ascending=False).head(1)

# mpg_df[mpg_df['manufacturer'] == 'dodge'] == mpg_df[mpg_df['manufacturer'] == 'dodge']
```

```python
# 11.2 Which dodge car has the worst average mileage?

mpg_df[mpg_df['manufacturer'] == 'dodge'].sort_values('average_mileage').head(4)
```

```python
#3.2. Summarize the dataframe with .describe.

m_df.describe()
```

```python
# 4. What is the the weight of the fastest animal?

m_df[m_df['speed'] == m_df['speed'].max()]['weight']

# m_df.sort_values('speed', ascending=False).head(1)
```

```python
# 5. What is the overall percentage of specials?

m_df.specials.sum() / m_df.specials.count() * 100
```

```python
# 6.1. How many animals are hoppers that are above the median speed?

m_df[m_df.hoppers & (m_df.speed > m_df.speed.median())].count()

# 7
```

```python
# 6.2. What percentage is this?

m_df.count() / m_df[m_df.hoppers & (m_df.speed > m_df.speed.median())].count()

# 15%
```
