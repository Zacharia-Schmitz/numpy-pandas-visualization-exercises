```python
# 1. Use matplotlib to plot the following equation:
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

# y = x^2 - x + 2

import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('dark_background')

x = pd.Series(range(-20, 20))
#y = [(x**2 - x + 2) for x in x]
y = x**2 - x + 2

plt.plot(x,y)

# add title
plt.title('$y = x^2 - x +2$')

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('Origin', xy = (0,0), xytext = (0, 200), arrowprops ={})

plt.show()
```

```python
# 2.1. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# You can use functions from the math module to help implement some of the equations.

# y = \sqrt{x}

import math

x = range(0,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')

plt.show()
```

```python
# 2.2 y = x^3

x = range(0, 20)
y = [n**3 for n in x]

plt.plot(x,y)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')

plt.show()
```

```python
# y = 2^x

x = range(1, 21)
y = [2**n for n in x]

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('$2^x$')
plt.title('$y=2^x$')

plt.show()
```

```python
# 2.4. y = 1/(x+1)

x = range(1,21)
y = [1/(n + 1) for n in x]

plt.plot(x,y)
plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')

plt.show()
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.

import matplotlib.pyplot as plt
import math

x = range(1,21)
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n + 1) for n in x]

plt.subplot(221)
plt.plot(x, y1)
plt.title('$\sqrt{x}$')

plt.subplot(222)
plt.plot(x, y2)
plt.title('$x^{3}$')

plt.subplot(223)
plt.plot(x, y3)
plt.title('$2^{x}$')
plt.ticklabel_format(style='plain') # Removes scientific notation

plt.subplot(224)
plt.plot(x, y4)
plt.title('$1/(x+1)$')

plt.suptitle('Math Equations')

plt.tight_layout(pad=1.0)

plt.ticklabel_format(style='plain')

plt.show()
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points.
# Be sure to include a legend and an appropriate title for the figure.

import matplotlib.pyplot as plt
import numpy as np

x = list(range(0, 25))

y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.plot(x, y1, c= 'red', label='$\sqrt{x}$')
plt.plot(x, y2, c= 'yellow', label='$x^{3}$')
plt.plot(x, y3, c= 'blue', label='$2^{x}$')
plt.plot(x, y4, c= 'green', label='$1/(x+1)$')

plt.ylim(0, 25)
plt.xlim(0, 25)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('My Functions')
plt.legend()

plt.show()
```

```python
from random import randint

# Some random data
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]

# we can use the alpha kwarg to define how transparent the color is
plt.plot(x1, c='green', alpha=0.6)
plt.plot(x2, c='red', alpha=0.4)
plt.show()
```
